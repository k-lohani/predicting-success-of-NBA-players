```python
import requests as rq
import pandas as pd
# !pip install nba_api
import nba_api
from nba_api.stats.endpoints import playercareerstats
import time
```

```python
pd.set_option('display.max_columns', None)
```

```python
season_list = ["1971-72","1970-71","1969-70","1968-69","1967-68","1966-67","1965-66","1964-65","1963-64","1962-63","1961-62","1960-61","1959-60","1958-59","1957-58","1956-57","1955-56","1954-55","1953-54","1952-53","1951-52","1950-51","1949-50","1948-49","1947-48","1946-47"]
# next
#  
```

```python
headers  = {
    'Connection': 'keep-alive',
    'Accept': 'application/json, text/plain, */*',
    'x-nba-stats-token': 'true',
    'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/79.0.3945.130 Safari/537.36',
    'x-nba-stats-origin': 'stats',
    'Sec-Fetch-Site': 'same-origin',
    'Sec-Fetch-Mode': 'cors',
    'Referer': 'https://stats.nba.com/',
    'Accept-Encoding': 'gzip, deflate, br',
    'Accept-Language': 'en-US,en;q=0.9',
}
```

```python
nba_dfs = []
for season in season_list:
  season_id = season[0:4]
  url = "https://stats.nba.com/stats/leagueLeaders?LeagueID=00&PerMode=PerGame&Scope=Rookies&Season="+season+"&SeasonType=Regular%20Season&StatCategory=PTS"
  res = rq.get(url = url, headers = headers).json()
  columns = res['resultSet']['headers']
  data = res['resultSet']['rowSet']
  nba_df = pd.DataFrame(data, columns = columns)
  nba_df['season'] = season_id
  print(season, ": ", nba_df.shape)
  nba_dfs.append(nba_df)
```

```python
nba_concat_df = pd.concat(nba_dfs)
display(nba_concat_df.head())
print(nba_concat_df.shape)
```

```python
nba_df_final = nba_concat_df[['PLAYER_ID','PLAYER', 'GP', 'MIN', 'PTS', 'FGM', 'FGA', 'FG3M', 'FG3A', 'FTM', 'FTA','OREB', 'DREB', 'AST', 'STL', 'BLK','TOV', 'season']]
```

```python
nba_df_final.head()
```

```python
new_cols = ['id','name','games_played','minutes_played','points_scored','goals_scored','goals_attempted','3_point_scored','3_point_attempted','free_throws_scored','free_throws_attempted','offensive_rebounds','defensive_rebounds','assists','steals','blocks','turnover', 'season']
nba_df_final.columns = new_cols
nba_df_final['5Yrs'] = -1
nba_df_final.head()
```

```python
nba_df_final.shape
```

```python
def extract_sea(sea_str):
  return int(sea_str[0:4])

def update_target(i):
  career = playercareerstats.PlayerCareerStats(player_id=i) 
  frames = career.get_data_frames()[0]
  frames['sea'] = frames['SEASON_ID'].apply(extract_sea)
  sea_max = max(frames['sea'])
  print(i)
  player_draft_season = int(nba_df_final[nba_df_final['id'] == i]['season'])
  if sea_max >= player_draft_season+5:
    nba_df_final.loc[nba_df_final['id'] == i, '5Yrs'] = 1
  else:
    nba_df_final.loc[nba_df_final['id'] == i, '5Yrs'] = 0
```

```python
for i in nba_df_final[nba_df_final['5Yrs'] == -1]['id'][0:429]:
  update_target(i)
```

```python
print("Total not updated Shape: ", nba_df_final[nba_df_final['5Yrs'] == -1].shape)

print("0 shape: ",nba_df_final[nba_df_final['5Yrs'] == 0].shape)

print("1 shape: ",nba_df_final[nba_df_final['5Yrs'] == 1].shape)
```
